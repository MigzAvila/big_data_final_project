```python
import pandas as pd
import os
import numpy as np
import glob
from IPython.display import display

```

```python
import os
os.environ["KAGGLE_CONFIG_DIR"] = "."

from kaggle.api.kaggle_api_extended import KaggleApi
api = KaggleApi()
api.authenticate()
```

```python
api.dataset_download_files(
    'anvitkumar/shopping-dataset',
    path='./data',                      
    unzip=True                       
)

print("Dataset downloaded successfully!")
```

```python
folder_path = "data/"

# Find all CSV files
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Dictionary to store dataframes separately
dfs_dict = {}

for file in all_files:
    df = pd.read_csv(file)
    file_name = os.path.basename(file)
    dfs_dict[file_name] = df


# Get a list of all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Read all CSVs into a list of dataframes
dfs = [pd.read_csv(f) for f in csv_files]

# Merge all dataframes on 'product_id', keeping all columns
# If there are duplicate product_ids, this will merge the rows
merged_df = pd.concat(dfs, ignore_index=True)

# Optional: drop exact duplicate rows (keeping first occurrence)
merged_df = merged_df.drop_duplicates(subset=['product_id'], keep='first')


```
